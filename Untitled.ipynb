## Some of the popular languages that Data Scientists use are:

1. Python - One of the most popular languages used by Data Scientists, Python is known for its simplicity and versatility. It has a rich collection of libraries such as Pandas, NumPy, and Scikit-learn, which make it suitable for various tasks such as data wrangling, machine learning, and visualization.

2. R - R is another popular language used by Data Scientists, especially in the field of statistical analysis. It has a wide range of statistical packages that make it easy to perform complex data analysis tasks.

3. SQL - While not a programming language per se, SQL (Structured Query Language) is an essential tool for Data Scientists as it provides a powerful way to manipulate and query data stored in databases. 

4. Julia - Julia is a relatively new language that has gained popularity among Data Scientists due to its speed and performance. It is designed for numerical and scientific computing and has a syntax similar to Matlab and Python.

5. Scala - Scala is a programming language that runs on the Java Virtual Machine (JVM). It is gaining popularity among Data Scientists due to its ability to handle big data and its interoperability with other JVM languages such as Java and Python.
